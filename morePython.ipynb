````python
class Car:
    #Constructor of this class
    def __init__(self, name, brand):
        self.name = name
        self.brand = brand

#Creating objects of this class
car1 = Car("Range Rover", "Land Rover") #REMEBER CONSTRUCTOR IS CALLED FIRST ALWAYS EVERYWHERE!!!

print(car1)
print(car1.brand)
print(car1.name)
````

````python
# define a class with instance methods

class Vechile:
    def __init__(self, carName, carBrand):
        self.carName = carName
        self.carBrand = carBrand
    
    def printCar(self): #we are mentioning self in her because it allows this method to then go ahead and use the variable we defined above in the Vechile class(carBrand and carName)
        print(f"This is Shivansh's Favourtie SUV: {self.carName} and it is manufactured by, {self.carBrand}")

car1 = Vechile("Range Rover", "Land Rover")
car1.printCar()
car2 = Vechile("Defender", "Land Rover")
car2.printCar()
````

````python
#Modeling a Bank Account

#define a class for the bank account
class BankAccount:
    def __init__(self, owner, balance = 0): # by default when the account is first created the balance is 0
        self.owner = owner
        self.balance = balance

    #deposit method
    def depositMoney(self, amount):
        self.balance = self.balance + amount
        print(f"{amount}, has been deposited. The new balance is {self.balance}")
    
    #withdraw method
    def withdrawMoney(self, amount):
        if amount > self.balance:
            print("This amount cannot be withdrawn")
        else:
            self.balance = self.balance - amount
            print(f"{amount}, has been withdrawed from the account. The current balance is {self.balance}")

    def getBalance(self):
        return self.balance
    
#Create an account

shivansh = BankAccount("Shivansh", 10)
print(shivansh.balance)
shivansh.depositMoney(90000000000000000000000)
shivansh.getBalance()
shivansh.withdrawMoney(350000)
shivansh.getBalance()
````

# Inheritance 

### In the cell below we only have one parent class so wecan just use super().__init__() but in the the case that we have one class that is inheriting from 
### multiple classes we would have to use the class name and then do the .__init__()
---
## Example:

```python
class X:
    def __init__(self, firstname):
        self.firstname = firstname

class Y:
    def __init__(self, lastname):
        self.lastname = lastname

class Z(X, Y):
    def __init__(self, firstname, lastname):
        X.__init__(self, firstname)
        Y.__init__(self, lastname)


````python
# Inheritance

#Parent Class
class Gadi:
    def __init__(self, windows, doors, engineType):
        self.windows = windows
        self.doors = doors
        self.engineType = engineType
    
    def drive(self):
        print(f"the person will drive the {self.engineType} car")

g1 = Gadi(4,5, "petrol")

#Child Class
class Tesla(Gadi):
    def __init__(self, windows, doors, engineType, isSelfDriving):
        super().__init__(windows, doors, engineType)
        self.isSelfDriving = isSelfDriving

    def selfDriving(self):
        print(f"The car is {self.engineType} and it can also drive by itself: {self.isSelfDriving}")
t1 = Tesla(4,4, "electric", True)
t1.selfDriving()
t1.drive()
````
