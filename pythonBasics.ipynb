# Shallow copy Vs Deep copy 


### Shallow Copy-> Shallow copy means uppar uppar kee copy karna. meaning if we have a list with certain elements like 1,2,3 but after 3 we have a another list inside the original list like lst=[1,2,3,[4,5]] and if we we do the .copy() method to copy tihs lst into lst1 then change the value of 1,2,3 in the lst1 to something else and print the both lists then the change will happen only in lst1 and lst will remain the same as defined originally. The problem occurs if we change the value of 4,5 in lst1 and then print both lists, the value of 4,5 will be changed in both to the new value in BOTH of the lists.


### Deep Copy->Deep Copy is basically solves the issue for Shallow Copy in the above example. It allows for both lists to be treated as two different lists and if we make change to one list it does not reflect the same change to both lists, only does change to the specified list.

````python
# Reference copy
l1 = {"name": "Shivansh", "age": 19, "state": "DE"}
l2 = l1

l2["name"] = "Shiv"
print(l1)
print(l2)
print()

# Shallow copy
l1 = {"name": "Shivansh", "age": 19, "state": "DE"}
l2 = l1.copy()

l2["name"] = "Shiv"
print(l1)
print(l2)
print()

# Deep copy
import copy

l1 = {"name": "Shivansh", "age": 19, "state": "DE"}
l2 = copy.deepcopy(l1)

l2["name"] = "Shiv"
print(l1)
print(l2)
````

# Python CSV Files — Writing & Reading Notes

---

## ✍️ Writing to a CSV File (One-Line Notes)

- `import csv` → Imports Python’s built-in module used to read and write `.csv` files.
- `csv` → Stands for *Comma Separated Values*, commonly used for tabular data (Excel, Kaggle).
- `open()` → Opens a file so Python can read from or write to it.
- `'example.csv'` → Name of the CSV file being created or accessed.
- `mode='w'` → Opens the file in write mode (creates a new file or overwrites an existing one).
- `mode='r'` → Opens the file in read mode.
- `mode='w+'` → Opens the file for both writing and reading.
- `newline=''` → Prevents extra blank lines when writing CSV files (especially on Windows).
- `with open(...) as file:` → Opens the file and automatically closes it after the block ends.
- `file` → File object representing the opened CSV file.
- `csv.writer(file)` → Creates a writer object used to write data into the CSV file.
- `writer` → Object responsible for writing rows into the CSV.
- `writer.writerow()` → Writes a single row to the CSV file.
- `["name", "age"]` → Header row defining column names.
- `["Shivansh", 19]` → Data row containing actual values.
- Each `writer.writerow()` call → Adds a new row to the CSV file.

---

## 📖 Reading from a CSV File (One-Line Notes)

- `open()` → Opens a file so Python can access its contents.
- `'example.csv'` → Name of the CSV file being read.
- `mode='r'` → Opens the file in read mode.
- `with open(...) as file:` → Opens the file and automatically closes it after use.
- `file` → File object representing the opened CSV file.
- `csv.reader(file)` → Creates a reader object that reads data from the CSV file.
- `reader` → Object that allows iteration over rows in the CSV file.
- `for row in reader:` → Loops through each row in the CSV file one at a time.
- `row` → A list containing values from one row of the CSV file.
- `print(row)` → Prints each row as a Python list.
- Each iteration of the loop → Reads and processes one row from the CSV file.


````python
#csv (we import this when we are dealing with .csv type of files (kaggle))
import csv

'''
This will create a new csv file called example.csv in write mode,
 - w -> write mode
 - r -> read mode
 - w+ -> write and read mode 
 - newline='' -> prevents extra blank lines in CSV (IMPORTANT on Windows)
'''
with open('example.csv', mode = 'w', newline= '') as file:
    writer = csv.writer(file)
    writer.writerow(["name", "age"])
    writer.writerow(["Shivansh", 19])

with open('example.csv', mode = 'r') as file:
    reader = csv.reader(file)
    for row in reader: 
        print(row)

````

# Python Regular Expressions (re) — Detailed Notes

---

## 🔹 What are Regular Expressions (Regex)?

- Regular expressions are **special patterns** used to **match, search, and manipulate text**.
- They allow you to **define rules** for strings, like "must contain a number", "starts with a capital letter", "contains an email", etc.
- Regex is widely used in **data cleaning, validation, searching logs, scraping text**, and many other applications.
- Example:  
  - Pattern `\d+` → matches **one or more digits**  
  - Pattern `[A-Za-z]+` → matches **one or more letters**  
  - Pattern `^Hello` → matches strings that **start with "Hello"**

---

## 🔹 What is the `re` module in Python?

- `re` is Python's **built-in library for regular expressions**.
- It provides functions to:
  - `search()` → find the **first match** of a pattern in a string
  - `match()` → check if the string **starts with the pattern**
  - `findall()` → find **all matches** of a pattern in a string
  - `sub()` → **replace** text that matches a pattern
  - `split()` → **split** a string by a regex pattern
- Using `re` allows **powerful string manipulation and pattern-based searching**.



# Python Regular Expressions (re) — One-Line Notes

- `import re` → Imports Python’s built-in module for working with regular expressions.
- `re` → Stands for *regular expressions*, used for pattern matching in strings.
- `pattern` → A regex pattern that defines what we want to search for in a string.
- `r'\d+'` → Regex pattern where `\d` means digit (0–9) and `+` means one or more occurrences.
- `r''` → Raw string notation that tells Python not to treat backslashes as escape characters.
- `text` → The string in which the pattern will be searched.
- `"I have 12 GPUs"` → Sample string containing both text and numbers.
- `re.search(pattern, text)` → Searches the string for the first occurrence of the pattern.
- `match` → Stores the match object returned if the pattern is found.
- `match.group()` → Returns the actual matched substring from the text.
- `print(match.group())` → Prints the matched pattern (`12`).
- If no match is found → `re.search()` returns `None`.


````python
#regular expressions
import re

#This program checks if a given string has a pattern we define (must contain number)

pattern = r'\d+' # d stands for digits. this patteern is used to search for numbers in a string
text = "I have 12 GPUs"
match = re.search(pattern, text)
print(match.group())
````

# Python File Open Modes (`open()`)

When opening a file in Python using `open(filename, mode)`, the **mode** determines
how the file is accessed (read, write, append, etc.).

---

## 📂 Basic File Modes

### `r` — Read
- Opens the file for **reading only**
- File **must exist**, otherwise an error occurs
- Cursor starts at the beginning

---

### `w` — Write
- Opens the file for **writing**
- **Creates a new file** if it does not exist
- **Overwrites existing content**

---

### `a` — Append
- Opens the file for **writing at the end**
- Creates file if it does not exist
- **Preserves existing content**

---

### `x` — Exclusive Create
- Creates a **new file only**
- Raises an error if the file already exists
- Prevents accidental overwrites

---

### `r+` — Read and Write
- Allows **both reading and writing**
- File must exist
- Cursor starts at the beginning
- Does **not** overwrite the file

---

### `w+` — Write and Read
- Allows reading and writing
- **Overwrites existing content**
- Creates file if it does not exist

---

## 🔧 Mode Modifiers

These can be combined with the modes above.

### `t` — Text Mode (default)
- Works with **text strings**
- Handles newline characters automatically

---

### `b` — Binary Mode
- Works with **bytes**
- Used for non-text files (images, audio, PDFs, etc.)

---

## 🔀 Common Mode Combinations

| Mode | Description |
|-----|-------------|
| `rb` | Read binary |
| `wb` | Write binary |
| `ab` | Append binary |
| `r+b` | Read and write binary |
| `w+t` | Write and read text |
| `a+` | Read and append |

---

## 📊 Mode Behavior Summary

| Mode | Read | Write | Create File | Overwrite |
|------|------|------|-------------|-----------|
| `r`  | ✅ | ❌ | ❌ | ❌ |
| `w`  | ❌ | ✅ | ✅ | ✅ |
| `a`  | ❌ | ✅ | ✅ | ❌ |
| `x`  | ❌ | ✅ | ✅ | ❌ |
| `r+` | ✅ | ✅ | ❌ | ❌ |
| `w+` | ✅ | ✅ | ✅ | ✅ |

# MEMORIZE THE CELL BELOW TWIN ITS IMPORTANT :)

````python
# read a text file and count the number of lines, words, and charachters

def countTextFile(file):
    with open(file, mode = 'r') as f:
        lines = f.readlines()
        lineCount = len(lines)
        wordCount = sum(len(line.split()) for line in lines)
        charCount = sum(len(line) for line in lines)
    return lineCount, wordCount, charCount

filePath = 'source.txt'
lines, words, chars = countTextFile(filePath)

print(f"Number of line: {lines}\n Number of words: {words}\n Number of Characters: {chars}")
````
