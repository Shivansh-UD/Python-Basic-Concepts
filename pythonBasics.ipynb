# Shallow copy Vs Deep copy 


### Shallow Copy-> Shallow copy means uppar uppar kee copy karna. meaning if we have a list with certain elements like 1,2,3 but after 3 we have a another list inside the original list like lst=[1,2,3,[4,5]] and if we we do the .copy() method to copy tihs lst into lst1 then change the value of 1,2,3 in the lst1 to something else and print the both lists then the change will happen only in lst1 and lst will remain the same as defined originally. The problem occurs if we change the value of 4,5 in lst1 and then print both lists, the value of 4,5 will be changed in both to the new value in BOTH of the lists.


### Deep Copy->Deep Copy is basically solves the issue for Shallow Copy in the above example. It allows for both lists to be treated as two different lists and if we make change to one list it does not reflect the same change to both lists, only does change to the specified list.

````python
# Reference copy
l1 = {"name": "Shivansh", "age": 19, "state": "DE"}
l2 = l1

l2["name"] = "Shiv"
print(l1)
print(l2)
print()

# Shallow copy
l1 = {"name": "Shivansh", "age": 19, "state": "DE"}
l2 = l1.copy()

l2["name"] = "Shiv"
print(l1)
print(l2)
print()

# Deep copy
import copy

l1 = {"name": "Shivansh", "age": 19, "state": "DE"}
l2 = copy.deepcopy(l1)

l2["name"] = "Shiv"
print(l1)
print(l2)
````

````python
#iterating over dictionary
l3 = {"name": "Shivansh", "age": 19, "state": "DE"}

#Iterating over keys
for i in l3.keys():
    print(i)
print()

#iterating over values
for i in l3.values():
    print(i)
print()

#iterating over items 
for i,j in l3.items(): #We can also just do i and it still gives the same output but the format is in the form of tuples
    print(f"{i}: {j}") # we do this to format how we are presenting the key and values.
````

````python
#Nested Dictionary

students = {
    "st1" : {"name": "Shivansh", "age": 19, "games":True},
    "st2" : {"name": "Cody", "age": 28, "games":True},
    "st3" : {"name": "Lacy", "age": 23, "games":False}
}

for i in students.values(): #first we iterate over the values in the dictonary associated with the keys(st1,st2,st3)
    print()
    for j, k in i.items(): # then since st1,st2,st3 are also dictionaries that have key value pairs, we can then iterate over them to get the keys or values inside them
        print(f"{j}: {k}")
print()

#changing the value inside
students["st1"]["name"]="Shiv"
print(students["st1"])
print()

for i in students.values():
    print(i)


    
    
````

````python
#Use a dictionary to count the frequence of elements in a list
numbers1 = [1,1,1,2,2,3,4,4,5,5,5,5,5,5,5]
frequency = {"Number of 1's": 0, "Number of 2's": 0, "Number of 3's": 0, "Number of 4's": 0, "Number of 5's": 0}
frequency2 = {}
#THIS IS NOT THE BEST WAY TO DO THIS BUT IT WORKS
for x in numbers1:
    if x == 1:
        frequency["Number of 1's"] +=1
    if x == 2:
        frequency["Number of 2's"] +=1
    if x == 3:
        frequency["Number of 3's"] +=1
    if x == 4:
        frequency["Number of 4's"] +=1
    if x == 5:
        frequency["Number of 5's"] +=1
print()

for i,j in frequency.items():
    print(f"{i}: {j}")

print("---------------------------------")
#BETTER WAY TO DO IT
for x in numbers1:
    if x in frequency2:
        frequency2[x] +=1
    else:
        frequency2[x] = 1

for i,j in frequency2.items():
    print(f"{i}: {j}")
````
